```python
variable = threes
data1 = pd.DataFrame(columns = ['Br11 EqW','Br20 Confidence','Br11 Confidence','Br20 EqW'])
br11 = []
br11confidence = []
br20confidence = []
br20eqw = []
for i in range(len(variable)):
    br11.append(variable[i][0])
    br11confidence.append(variable[i][1])
    br20confidence.append(variable[i][2])
    br20eqw.append(variable[i][3])
br11 = np.asarray(br11)
br20confidence = np.asarray(br20confidence)
br11confidence = np.asarray(br11confidence)
br20eqw = np.asarray(br20eqw)
data1['Br11 EqW'] = br11
data1['Br20 Confidence'] = br20confidence
data1['Confidence'] = br11confidence
data1['Br20 EqW'] = br20eqw
```

## Br20 Confidence vs. Br20 EqW 

```python
p = sns.JointGrid(y='Br20 Confidence',x='Br20 EqW', data=data1, size=10);

p = p.plot_joint(plt.scatter,s=5,color='black');
plt.xlim(-2.5,7.5)
p = p.plot_joint(sns.kdeplot,shade=True,shade_lowest=False,cmap='RdGy',zorder=0,n_levels=15)

p = p.plot_marginals(sns.distplot,kde=True)

plt.suptitle('Threes')
```

```python
plt.figure(figsize=(13,10))
sns.distplot(br20confidence,bins=20,kde=False,norm_hist=False,kde_kws={'color':'black','ls':'dashed','lw':2},
            hist_kws={'edgecolor':'black','linewidth':1,'color':'blue','alpha':.7})
plt.xlabel('Fmax/Continuum Error')
plt.title('Threes')
```
