```python
def superplotter2(x,temp,density):
    plt.figure(figsize=(20,10));
    for column in x:
        g = x['Br Line']
        if column != 'Br Line':
            y = x[column]
            y = y/y[0]
            plt.plot(g,y,label=column+' Log Density');
            plt.scatter(g,y,label='_nolabel_');
            plt.title(temp + ' Ratios',fontsize=20);
            plt.xlabel('n$_{up}$',fontsize=20);
            plt.ylabel('Log H$_{nup}/H_{4}$',fontsize=20);
            plt.grid(True,which='both')
            plt.xticks(np.arange(11,21,1));
    plt.legend(bbox_to_anchor=(1,1),fontsize=15);
    plt.savefig(savefile+temp+density+'.pdf',bbox_inches='tight',dpi=300)
    plt.close()
```
