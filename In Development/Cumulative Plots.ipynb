# New Plots

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
data=[]
above=[]
with open('All Visits.csv') as csvfile:
        
    reader = csv.DictReader(csvfile,delimiter='\t')
    
    for row in reader:
        data.append(float(row['Equivalent Width']))
data.sort()

for i in range(len(data)):
    if float(data[i])>1.200000 and float(data[i])< 50.0:
        above.append(data[i])
above.sort()
```

```python
# evaluate the histogram
#binwidth = 1
#values, base = np.histogram(data, bins=np.arange(min(data), max(data) + binwidth, binwidth))
#evaluate the cumulative

#cumulative = np.cumsum(data)
#total_percent = len(cumulative)/len(data)

# plot the cumulative function
plt.hist(above,bins=100000,normed=True,cumulative=True,histtype='step',log=True)

#plot the survival function
#plt.plot(data, len(data)-cumulative, c='green')

plt.xlim(1.15,2.5)
#plt.ylim(0.95,1.)
plt.xlabel('Equivalent Width')
plt.ylabel('% of Sample')

plt.savefig('Cumulative.pdf',dpi=300)
plt.show()
```

```python
datax=[]
datay=[]
with open('All Visits.csv') as csvfile:
        
    reader = csv.DictReader(csvfile,delimiter='\t')
    
    for row in reader:
        if float(row['Equivalent Width']) >0 and float(row['Standard Deviation'])>0:
            datax.append(float(row['Equivalent Width']))
            datay.append(float(row['Standard Deviation']))
```

```python
plt.xscale('log')
plt.yscale('log')
plt.scatter(datax,datay,s=0.1)
plt.xlabel('Equivalent Width')
plt.ylabel('Standard Deviation')
#plt.xlim(0,50)
plt.savefig('EqWStd Scatter Plot.pdf',dpi=300)
plt.show()
```

```python
dataxx=[]
datayy=[]
with open('Average Visits.csv') as csvfile:
        
    reader = csv.DictReader(csvfile,delimiter='\t')
    
    for row in reader:
        if float(row['Average Equivalent Width'])>1.2 and float(row['Average Equivalent Width'])<50:
            dataxx.append(float(row['Average Equivalent Width']))
            datayy.append(float(row['SNR']))
```

```python
plt.xscale('log')
plt.yscale('log')
plt.scatter(dataxx,datayy,s=0.1)
plt.xlabel('EqW')
plt.ylabel('SNR')
plt.xlim(0,50)
#plt.ylim(0,2000)
plt.savefig('EqW and SNR.pdf',dpi=300)
plt.show()
```

```python
plt.hist(datayy,bins=100000,normed=True,cumulative=True,histtype='step',log=True)
plt.xlabel('SNR')
plt.ylabel('% of sample')
plt.xlim(-10,1000)
plt.savefig('SNR Cumulative.pdf',dpi=300)
```
