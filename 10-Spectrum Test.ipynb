## 10-Spectrum Test

```python
import apogee.tools.read as apread
import apogee.spec.plot as splot
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
Lambda = splot.apStarWavegrid()
```

```python
#Read in APOGEE Spectrum
visit1 = 1
spec10= apread.apStar(4586,'2M03434449+3143092',ext=1,header=False)[1 + visit1]
spec11= apread.apStar(4586,'2M03434449+3143092',ext=1,header=False)[2 + visit1]
spec12= apread.apStar(4586,'2M03434449+3143092',ext=1,header=False)[4 + visit1]
spec13= apread.apStar(4586,'2M03434449+3143092',ext=1,header=False)[8 + visit1]
spec14= apread.apStar(4586,'2M03434449+3143092',ext=1,header=False)[16 + visit1]
spec2= apread.apStar(4587,'2M03250943+3046215',ext=1,header=False)[1 + visit1]
spec3= apread.apStar(4587,'2M03292187+3115363',ext=1,header=False)[1 + visit1]
spec4= apread.apStar(4587,'2M03295403+3120529',ext=1,header=False)[1 + visit1]
spec5= apread.apStar(4572,'2M06413876+0932117',ext=1,header=False)[1 + visit1]
spec6= apread.apStar(4572,'2M06410890+0929451',ext=1,header=False)[1 + visit1]
spec7= apread.apStar(4572,'2M06410715+0927294',ext=1,header=False)[1 + visit1]
spec8= apread.apStar(4572,'2M06405926+0933250',ext=1,header=False)[1 + visit1]
spec9= apread.apStar(4572,'2M06405639+0935533',ext=1,header=False)[1 + visit1]
spec20= apread.apStar(4572,'2M06405426+0949203',ext=1,header=False)[1 + visit1]
spec21= apread.apStar(4572,'2M06404837+0948385',ext=1,header=False)[1 + visit1]
```

### Spec 1

```python
#Spec10
#Calculate Fc
lsum= np.sum(spec10[7545:7590])/ len(spec10[7545:7590])
rsum = np.sum(spec10[7719:7762])/len(spec10[7719:7762])
Fc= (lsum+rsum)/2
#Calculate Equivalent Width
EqW=0
for i in range(7590,7719):
    summ=(Fc*(Lambda[i+1]-Lambda[i]))-((1./2.)*(Lambda[i+1]-Lambda[i])*(spec10[i+1]+spec10[i]))
    EqW = EqW + summ
EqW = abs(EqW/Fc)
print EqW
upper = Lambda[7650]+(EqW/2)
lower = Lambda[7650]-(EqW/2)
#Plot averaged spectrum with EqW
fig = plt.figure()
plt.plot(Lambda,spec10)
plt.axhline(y=Fc,ls='dashed',color='black')
#plt.axvline(x=16760,ls='dashed',color='purple')
#plt.axvline(x=16770,ls='dashed',color='purple')
#plt.axvline(x=16800,ls='dashed',color='purple')
#plt.axvline(x=16810,ls='dashed',color='purple')
plt.axvline(x=Lambda[7650],ls='dashed',color='r')
#plt.axvline(x=16801,ls='dashed',color='r')
plt.axvspan(lower,upper,ymin=0,ymax= 0.5,facecolor='g',alpha=0.35)
fig.suptitle('2M03434449+3143092 EqW', fontsize=20)
plt.xlabel('Wavelength', fontsize=16)
plt.ylabel('Flux', fontsize=16)
plt.xlim([16750,16820])
plt.ylim([0,Fc*2])
plt.savefig('EqW Test_1.jpg',bbox_inches='tight')
```
