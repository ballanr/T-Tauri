```python
#Recreating reading in and plotting from the APOGEE tool documentation
#Mainly just to make sure python is up and running with various packages

import apogee.tools.read as apread
import apogee.spec.plot as splot
from matplotlib.pyplot import draw,show
%matplotlib inline
data = apread.rcsample()
indx= data['SNR'] > 200.
data= data[indx]

splot.waveregions(data[3513]['LOCATION_ID'],data[3513]['APOGEE_ID'],ext=1,
                  labelID=data[3513]['APOGEE_ID'],
                  labelTeff=data[3513]['TEFF'],
                  labellogg=data[3513]['LOGG'],
                  labelmetals=data[3513]['METALS'],
                  labelafe=data[3513]['ALPHAFE'])
```

```python
#Now I'm going to try a file walk courtesy of code from Jessica
import numpy as np
import os
```

```python
infile = np.loadtxt('master_apStarList.lis',dtype='str',usecols=[2])
topdir='/Volumes/CoveyData/APOGEE_Spectra/APOGEE_DR12_ApStar'
exten='.fits'
```

```python
outfile=open('APOGEE_List.lis','w')
for dirpath,dirnames,files in os.walk(topdir):
    for i in files:
        if i.lower().endswith(exten):
            path = os.path.join(dirpath,i)
            path = path[-38:]
            outfile.write(str(path)+'\n')
                
```
